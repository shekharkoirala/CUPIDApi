```python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"
model_name = "gpt-4o"

client = OpenAI(
    base_url=endpoint,
    api_key=token,
)

```

```python
import json
import random
import time
# print(normalize_room_name("Deluxe Suite with Balcony!"))  
with open("../data/merged.json", "r") as f:
    data = json.load(f)
```

```python
check  = []
for d in data:
    if set(d["reference"]) != set(d["supplier"]):
        check.append(d)

len(check)
```

```python
for d in check[30:60]:
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": 'You work with Hotel API Management company, You know How to exactly match two room name with its feature. There is no room for error. You will only output the matched cases when they are similar in most of cases, there is only difference in name like no pets allowded or no smoking, but you should not ignore fact like 1bed room to 2 bed room. Give match result like :  Match = [{"supplier_room_name" : "superior room score","reference_room_name" : "superior room ada room only"}] and ignore all unmatched cases and also ignore the same names like this: {"supplier_room_name": "Deluxe Room", "reference_room_name": "Deluxe Room"}, just send the match cases. no other text',
            },
            {
                "role": "user",
                "content": f"Find the matched room name from the list of data: {d} ",
            }
        ],
        temperature=1.0,
        top_p=1.0,
        max_tokens=1000,
        model=model_name
    )

    print(response.choices[0].message.content)
    time.sleep(1)
```
